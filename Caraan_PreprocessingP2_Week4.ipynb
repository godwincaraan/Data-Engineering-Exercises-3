```python
missing_data = pd.read_csv('MissingValueEx.csv')
missing_data
```

```python
missing_data.isnull().sum()
```

#### I intend to retain all columns within this dataset, as the quantity of missing values is not substantial enough to warrant their removal.

```python
#getting the mode
mode_missing_data = missing_data['Sex'].mode()

#filling the NaN values with the mode value
missing_data['Sex'] = missing_data['Sex'].fillna(mode_missing_data[0])
missing_data
```

#### I decided to employ the mode to complete the missing values in the 'Sex' category, instead of resorting to 'No Data'. This choice was made because 'Sex' holds significant information that can be interpreted meaningfully and add value to the dataset.

```python
# getting the mean
mean_missing_data = missing_data[['Income','Children']].mean()

# filling the missing data using mean
missing_data[['Income','Children']] = missing_data[['Income','Children']].fillna(mean_missing_data).round(0) #include rounding
missing_data
```

#### I opted for utilizing the mean to impute the absent values, as opposed to the median, given that the dataset follows a normal distribution.

```python
#updating the file
missing_data.to_csv('MissingValueEx.csv',index = False)

updated_missing_data = pd.read_csv('MissingValueEx.csv')
updated_missing_data
```

```python
import pandas as pd
import numpy as np
```

```python
commuters_data = pd.read_csv('commuters.csv')
commuters_data
```

```python
#check if there is data duplication
commuters_data.duplicated().any()
```

```python
#remove the duplicate data
commuters_data = commuters_data.drop_duplicates()
commuters_data
```

```python
commuters_data.isnull().sum()
```

```python
#drop the name column since it has many null values
commuters_data = commuters_data.drop('Name',axis=1)
```

```python
commuters_data
```

```python
# Renaming the column of travel time
commuters_data.rename(columns={"Travel_Mins": "Travel_Hrs"}, inplace=True)

#converting mins to hours
commuters_data['Travel_Hrs'] = commuters_data['Travel_Hrs'] / 60

# Renaming the column of travel distance
commuters_data.rename(columns={"Distance_M": "Distance_KM"}, inplace=True)

#converting meters to km
commuters_data['Distance_KM'] = commuters_data['Distance_KM'] / 1000
```

```python
commuters_data
```

```python
#insert new column
commuters_data.insert(5,'Standardized_Distance_KM', 0)
```

```python
#standardization using the formula
commuters_data['Standardized_Distance_KM'] = (commuters_data['Distance_KM'] - np.mean(commuters_data['Distance_KM'])) / np.std(commuters_data['Distance_KM'])

# Set a z-score threshold (e.g., 2)
z_distance_threshold = 2

# Identify outliers based on threshold
outliers = commuters_data[abs(commuters_data['Standardized_Distance_KM']) > z_distance_threshold]
outliers
```

```python
# Remove outliers from the dataset
cleaned_commuters_data = commuters_data[abs(commuters_data['Standardized_Distance_KM']) <= z_score_threshold]
cleaned_commuters_data
```

```python
# Getting the mode
mode_gender = cleaned_commuters_data['Gender'].mode()

# Filling in the missing data with the mode
cleaned_commuters_data['Gender'] = cleaned_commuters_data['Gender'].fillna(mode_gender[0])
```

```python
cleaned_commuters_data
```

```python
# Getting the mean
mean_expenses = cleaned_commuters_data['Daily_Expenses_Php'].mean()

# Filling in the missing data with the mean
cleaned_commuters_data['Daily_Expenses_Php'] = cleaned_commuters_data['Daily_Expenses_Php'].fillna(mean_expenses).round(0)
```

```python
cleaned_commuters_data
```
